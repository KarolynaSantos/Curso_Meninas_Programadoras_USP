```python
def binary_extraction_par(lista, elementos, meio):

  lista_final = []
  lista_final.append(lista[meio])

  elemento = 1

  while True:
    if len(lista_final) == elementos:
      break

    lista_final.append(lista[meio - elemento])

    if len(lista_final) == elementos:
      break

    lista_final.append(lista[meio + elemento])

    elemento +=1

  return lista_final

def binary_extraction_impar(lista, elementos, meio):

  lista_final = []
  lista_final.append(lista[meio])

  elemento = 1

  while True:
    if len(lista_final) == elementos:
      break

    lista_final.append(lista[meio + elemento])

    if len(lista_final) == elementos:
      break

    lista_final.append(lista[meio - elemento])

    elemento +=1

  return lista_final

def binary_extraction(entrada, elementos):

  lista_inicial = [int(x) for x in entrada.split(' ')]

  meio = int(len(lista_inicial) / 2)
  if elementos == 0:
    resultado = []
  elif len(lista_inicial) % 2 == 0:
    resultado = binary_extraction_par(lista_inicial, elementos, meio)
  else:
    resultado = binary_extraction_impar(lista_inicial, elementos, meio)

  print(lista_inicial)
  print(resultado)

# recebendo a lista
# entrada = input()

valores = input()
elementos = int(input())
binary_extraction(valores, elementos)
```

```python
def extrai_elementos_centrais(lista, n):
    tamanho = len(lista)

    # Verifica se o número n é maior que a metade do tamanho da lista
    if n > tamanho // 2:
        n = tamanho // 2

    # Extrai os elementos centrais da lista
    elementos_centrais = lista[tamanho // 2 - n : tamanho // 2 + n]

    return elementos_centrais

# Lê a lista de inteiros
entrada = input()
lista = list(map(int, entrada.split()))

# Lê o número inteiro
n = int(input())

# Verifica se o número está dentro dos limites da lista
if 0 <= n <= len(lista):
    # Extrai os elementos centrais
    resultado = extrai_elementos_centrais(lista, n)

    # Imprime a lista original e os elementos centrais
    print(resultado)
else:
    print("O número inteiro deve estar entre 0 e o tamanho da lista.")
```

```python
#Recebendo o texto
verso = input().lower()

#Ignorando a pontuação e Colocando em minusculos

pontuacao = [';','"',"'",'.',',']

for i in pontuacao:
  verso = verso.replace(i, '')

#Quebrando
palavras = verso.split(" ")

dicionario = {}

# Contando as palavras
for palavra in palavras:
    if palavra in dicionario:
        dicionario[palavra] += 1 #se está presente, conta =+1 na palavra
    else:
        dicionario[palavra] = 1 #se não existir é igual a 1

print(dicionario)

#recebendo a solicitação do usuario

while True:
    validando = input().lower()
    if validando =='.':
        break
    if validando in dicionario:
        resultado = dicionario[validando]
        print(resultado)
    else:
        print('%s não encontrada' % validando)
```
